```python
import pandas as pd
import numpy as np

file_path = r"C:\Users\Chris\Downloads\auto+mpg\auto-mpg.data"

# Define the column names
columns = ["mpg", "cylinders", "displacement", "horsepower", "weight", 
           "acceleration", "model_year", "origin", "car_name"]

# Load the dataset 
data = pd.read_csv(file_path, delim_whitespace=True, names=columns)

# Replace '?' with NaN 
data['horsepower'] = data['horsepower'].str.strip()
data['horsepower'] = data['horsepower'].replace('?', np.nan)

# Convert the 'horsepower' column to numeric
data['horsepower'] = pd.to_numeric(data['horsepower'])

#Checking to see NaN values
print(data.head(50))
```

```python

file_path = r"C:\Users\Chris\Downloads\auto+mpg\auto-mpg.names"
# Open the file and read its contents
with open(file_path, 'r') as file:
    contents = file.read()
# Print the contents of the file
print(contents)
```

```python
# Replace number values with continents and USA
data['origin'] = data['origin'].replace({1: 'USA', 2: 'Asia', 3: 'Europe'})

# testing to see if origin replaced the labels
print(data[['origin']].head(20))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.countplot(x='cylinders', data=data)
plt.title('Distribution of Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatterplot for horsepower vs. weight
plt.figure()
sns.scatterplot(x='horsepower', y='weight', data=data)
plt.title('Relationship Between Horsepower and Weight')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#weight vs acceleration
plt.figure()
sns.scatterplot(x='weight', y='acceleration', data=data)
plt.title('Are lighter cars generally faster?')
plt.xlabel('Weight')
plt.ylabel('Acceleration')

# Add a trendline
sns.regplot(x='weight', y='acceleration', data=data, scatter=False, color='red')
plt.grid(True)
plt.show()

```
