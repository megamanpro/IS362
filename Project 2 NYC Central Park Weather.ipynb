```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Chris\Downloads\NYC_Central_Park_weather_1869-2022.csv')
print(df)
```

```python
#clearing out the NaN values and replacing it with 0
df = df.fillna(0)
df
```

```python
#renaming the column names for clarity
df.rename(columns={
    'PRCP': 'Precipitation (inches)',
    'SNOW': 'Snowfall (inches)',
    'SNWD': 'Snow Depth (inches)',
    'TMIN': 'Minimum Temperature (°F)',
    'TMAX': 'Maximum Temperature (°F)'
}, inplace=True)
df
```

```python
#Checking to see if all maximum temperatures are actually greater than the minimum temperatures 
inconsistent_temp = df[df['Maximum Temperature (°F)'] < df['Minimum Temperature (°F)']]
print(inconsistent_temp)
#Swapping the values of the min and max with each other so that they're true 
df.loc[df['Minimum Temperature (°F)'] > df['Maximum Temperature (°F)'], ['Minimum Temperature (°F)', 'Maximum Temperature (°F)']] = \
    df.loc[df['Minimum Temperature (°F)'] > df['Maximum Temperature (°F)'], ['Maximum Temperature (°F)', 'Minimum Temperature (°F)']].values


```

```python
# Calculate Temperature Range
df['Temperature Range (°F)'] = df['Maximum Temperature (°F)'] - df['Minimum Temperature (°F)']

# Calculate Average Temperature
df['Average Temperature (°F)'] = (df['Maximum Temperature (°F)'] + df['Minimum Temperature (°F)']) / 2

df
```

```python
#Checking to see if there's any outlier values
print(df.describe())
```

# NYC Central Park Weather Analysis
This database analyzes historical weather data for NYC. I replaced all the NaN values with 0, checked if there was any issues with min being larger than max temperatures and swaped those values so that it would be true. Used describe to see if there was any other outlier data. 
Obversation we can make from this set, Central Park gets rain less than 50% of the days and snow less than 75% of the days. There's also a huge range in temperature, with a min of -15 and a max of 106, which shows a high variance in the temperature in NYC.
