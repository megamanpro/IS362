```python
import numpy as np    
import pandas as pd   
from pandas import DataFrame, Series  

flights_cols =['Company', 'Status', 'Los Angeles',	'Phoenix',	'San Diego'	,'San Francisco',	'Seattle']
flights = pd.read_csv(r'C:/project 1/Alaska and AM West Flight delays.csv',sep=',', names=flights_cols)
#removing bad data from graph like improper rows or NaN values
flights.drop(index=0, inplace=True) 
flights.drop(index=3, inplace=True) 
flights.loc[[2], 'Company'] = 'ALASKA'
flights.loc[[5], 'Company'] = 'AM WEST'
flights.index = range(1, len(flights) + 1)
flights


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
cities = ['Los Angeles', 'Phoenix', 'San Diego', 'San Francisco', 'Seattle']
alaska_delayed = [62, 12, 20, 102, 305]
amwest_delayed = [117, 415, 65, 129, 61]

# Create bar graph
x = np.arange(len(cities))
width = 0.20

#Set the size of the bars and color
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, alaska_delayed, width, label='Alaska', color='skyblue')
bars2 = ax.bar(x + width/2, amwest_delayed, width, label='AM West', color='lightcoral')

# Add labels, title, and legend
ax.set_xlabel('Cities')
ax.set_ylabel('Number of Delayed Flights')
ax.set_title('Delayed Flights in 5 Cities (Alaska vs. AM West)')
ax.set_xticks(x)
ax.set_xticklabels(cities)
ax.legend()
```

In Los Angeles, AM West had almost twice the number of delayed flights at 117, compared to Alaska's 62.
In Phoenix, the difference in delays is night and day. AM West had a staggering 415 delayed flights, while Alaska had only 12. This city is also Alaska's lowest delayed flight count.
In San Diego, numbers are relativitely low for both companies but Alaska still has less delays.
In San Francisco, both company have around the same amount of delays with each other here, potentially coming across similar issues.
In Seattle, this is the only city where Alaska overwhelmingly has more delays than AM West at 305 versus 61. 

Conclusions: AM West generally has more delays than Alaska, showing that as a company, they are dealing with more issues that are causing delays. Alaska is generally good with their timing except in Seattle. If they are able to fix their problem specifically in Seattle, they would be great across the board.
