```python
import pandas as pd   
import matplotlib.pyplot as plt

#Series showing the amount of miles we ended each day with
miles_eod = pd.Series([200,450,780,900,1200,1300,1434],index=["Day 1", "Day 2", "Day 3", "Day 4", "Day 5", "Day 6", "Day 7"])
#subtracts the current day miles from the previous one
daily_miles = miles_eod - miles_eod.shift(1)
#accounts for [0] trying to subtract from nothing or else it results in NaN/just manually sets it to the value of the first day
daily_miles.iloc[0] = miles_eod.iloc[0]
#Change the type from float to int for consistancy
daily_miles = daily_miles.astype(int)
#Create a dataframe with both total miles and daily miles shown
df = pd.DataFrame({
    "Miles Done Today": daily_miles,
    "Total Miles": miles_eod
})
df
```

```python
#Creates a blank graph
plt.figure();
#Uses our default index values for x and miles for y to plot the data
df.plot();
```
